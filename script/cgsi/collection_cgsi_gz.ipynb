```python
headers = {
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Encoding': 'gzip, deflate, br, zstd',
    'Accept-Language': 'zh-CN,zh;q=0.9,en;q=0.8',
    'Connection': 'keep-alive',
    'Cookie': 'JSESSIONID=F80B3BBC1029874906E76C8FF755E3F0; Hm_lvt_a6039968c788e754656d429e62818ff4=1712729510,1712819071,1712884038; Hm_lvt_bd536ccff249ff9cec8e28130c46d528=1712729510,1712819071,1712884038; Hm_lpvt_a6039968c788e754656d429e62818ff4=1712903165; Hm_lpvt_bd536ccff249ff9cec8e28130c46d528=1712903165',
    'Host': 'search.ndcp.cgsi.cn',
    'Sec-Ch-Ua': '"Google Chrome";v="123", "Not:A-Brand";v="8", "Chromium";v="123"',
    'Sec-Ch-Ua-Mobile': '?0',
    'Sec-Ch-Ua-Platform': '"Windows"',
    'Sec-Fetch-Dest': 'iframe',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-User': '?1',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
}

def get_data(dataid):
    url = f'https://search.ndcp.cgsi.cn/swzxDrill/map/markerInfo?lx=undefined&id={dataid}'
    response = requests.get(url,headers=headers,verify=False)
    # 假设这是您通过requests库获取到的HTML文本
    html_text = response.text  # 确保response.text是从之前的请求获取的HTML文本
    # 解析HTML文本
    soup = BeautifulSoup(html_text, 'html.parser')

    # 提取数据
    data = {
        "钻孔编号": soup.find("td", string="钻孔编号：").find_next_sibling("td").div.text.strip(),
        "钻孔类型": soup.find("td", string="钻孔类型：").find_next_sibling("td").div.text.strip(),
        "钻孔深度": soup.find("td", string="钻孔深度(m)：").find_next_sibling("td").div.text.strip(),
        "实物档号": soup.find("td", string="实物档号：").find_next_sibling("td").div.text.strip(),
        "项目名称": soup.find("td", string="项目名称：").find_next_sibling("td").div.text.strip(),
        "保管单位": soup.find("td", string="保管单位：").find_next_sibling("td").div.text.strip(),
        "数字化图形": {},
        "详情链接": soup.find("td", align="right").find("a")['href']
    }

    # 针对数字化图形获取所有链接
    digital_graphics_td = soup.find("td", string="数字化图形：").find_next_sibling("td")
    links = digital_graphics_td.find_all('a')
    for link in links:
        if 'javascript:void(0);' in link['href']:
            # 正则表达式搜索目标URL
            scripts = soup.find_all('script')
            for script in scripts:
                if 'openLayerByCfg' in script.text:
                    result = re.search(r'url:"([^"]+)"', script.text)
                    if result:
                        data["数字化图形"][link.text.strip()] = 'https://search.ndcp.cgsi.cn' + result.group(1)
                        break
        else:
            data["数字化图形"][link.text.strip()] = link['href']

    return data
```
